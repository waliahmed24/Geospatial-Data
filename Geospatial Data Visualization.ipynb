```python
#importing essential libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mplleaflet
```

```python
#reading the csv file containg latitude and longitude data and storing it into a dataframe
file = 'uber-raw-data-apr142.csv'
dfmin = pd.read_csv(file)
```

```python
dfmin.head()
```

```python
dfmin.shape
```

```python
#dropping unnecessary columns
cols = ['Unnamed: 0', 'Base']
dfmin = dfmin.drop(cols, axis = 1)
```

```python
dfmin.head()
```

```python
dfmin.dtypes
```

```python
#No null values found in the data
dfmin.isna().sum()
```

```python
dfmin.describe(include="all")
```

```python
#sampling data to reduce size
dfmin = dfmin.sample(frac=0.2)
dfmin.shape
```

```python
#plot lat lon in data through matplotlib
fig = plt.figure()
plt.plot(dfmin['Lon'].values, dfmin['Lat'].values, 'rs')
```

```python
#project matplotlib's graph onto mplleaflet map
mplleaflet.display(fig=fig)

#use mplleaflet.show(fig=fig) to display the map in a separate window
```
